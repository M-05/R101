````r
install.packages("ggplot2")
library(ggplot2)
````

````r
# Motor Trend Car Road Tests

ggplot(data=mtcars, aes(x=wt, y=mpg)) +
  geom_point() + # geom 함수 create scatterplots ? geometric ?
  labs(x="Weight (1,000 lbs)", # 제목 캡션 축이름 등의 그래프 부가 요소 ? labels?
       y="Fuel Consumption (miles per gallon)",
       title="Fuel Consumption vs Weight",
       subtitle="Negative relationship between fuel efficiency and car weight",
       caption="Source: mpg dataset")
````

````r
data(mtcars)

mtcars$cyl = factor(mtcars$cyl,
                    levels=c(4, 6, 8),
                    labels=c("4 cylinders", "6 cylinders", "8 cylinders"))
````

````r
ggplot(data=mtcars, aes(x=mpg)) +
  geom_histogram() +
  facet_grid(cyl ~ .) +
  labs(title="geom_histogram()",
        x="Miles per Gallon")
````

````r
ggplot(data=mtcars, aes(x=cyl, y=mpg)) +
  geom_boxplot() +
  labs(title="geom_boxplot()",
        x="Number of Cylinders",
        y="Miles per Gallon")
````

````r
ggplot(data=mtcars, aes(x=mpg, fill=cyl)) +
  geom_density() +
  labs(title="geom_desnity",
        x="Miles per Gallon")
````

````r
ggplot(data=mtcars, mapping=aes(x=wt, y=mpg, col=cyl)) +
  geom_point() +
  labs(title="geom_point()",
        x="Weight (1,000 lbs)",
        y="Miles per Gallon")
````

````r
ggplot(data=mtcars, mapping=aes(x=wt, y=mpg)) +
  geom_smooth() +
  labs(title="geom_smooth()",
        x="Weight (1,000 lbs)",
        y="Miles per Gallon")
````
