````r
library(ggplot2)
# Motor Trend Car Road Tests.f
head(mtcars)
tail(mtcars)
````

````r
ggplot(data=mtcars, mapping=aes(x=wt, y=mpg)) +
  geom_point() +
  geom_smooth() +
  labs(title="Fuel consumption vs Weight",
        x="Weight (1,000 lbs)",
        y="Fuel Consumption (miles per gallon)")
````

````r
ggplot(data=mtcars, mapping=aes(x=wt, y=mpg)) +
  geom_point(shape=11, color="pink", size=5.5) +
  geom_smooth() +
  labs(title="Fuel consumption vs Weight",
        x="Weight (1,000 lbs)",
        y="Fuel Consumption (miles per gallon)")
````

````r
ggplot(data=mtcars, mapping=aes(x=wt, y=mpg)) +
  geom_point(shape=21, size=5.5, stroke=2, # 두께
             color="chocolate", bg="skyblue") +
  geom_smooth() +
  labs(title="Fuel consumption vs Weight",
        x="Weight (1,000 lbs)",
        y="Fuel Consumption (miles per gallon)")
````

````r
?geom_smooth
````

````r
ggplot(data=mtcars, mapping=aes(x=wt, y=mpg)) +
  geom_point(shape=21, size=5.5, stroke=2, # 두께
             color="chocolate", bg="skyblue") +
  geom_smooth(method="lm", color='red') +
  # ‘"lm"’, ‘"glm"’, ‘"gam"’, ‘"loess"’
  labs(title="Fuel consumption vs Weight",
        x="Weight (1,000 lbs)",
        y="Fuel Consumption (miles per gallon)")
````

````r
ggplot(data=mtcars, mapping=aes(x=wt, y=mpg)) +
  geom_point(shape=21, size=5.5, stroke=2, # 두께
             color="chocolate", bg="skyblue") +
  geom_smooth(method="lm", color='red', linetype=3, size=2) +
  labs(title="Fuel consumption vs Weight",
        x="Weight (1,000 lbs)",
        y="Fuel Consumption (miles per gallon)")
````

````r
ggplot(data=mtcars, mapping=aes(x=wt, y=mpg)) +
  geom_point() +
  geom_smooth(method="glm", color='blue', linetype=2, size=2) +
  geom_text(label=rownames(mtcars),
            vjust=0, hjust=0, nudge_y=0.7,
            size=3.5) +
  labs(title="Fuel consumption vs Weight",
        x="Weight (1,000 lbs)",
        y="Fuel Consumption (miles per gallon)")
````

````r
install.packages("car")

library(car)
str(Salaries)
````

````r
ggplot(Salaries, aes(x=rank, y=salary)) +
  geom_boxplot()
````

````r
ggplot(Salaries, aes(x=rank, y=salary)) +
  geom_boxplot(fill="salmon", color="dimgray", notch=T)
# 홈이 파인 곳의 길이가 신뢰구간 95%
# 3개의 직급의 홈이 어디에도 겹치지 않기에
# 통계적으로 차이가 있다.
````

````r
ggplot(Salaries, aes(x=rank, y=salary)) +
  geom_boxplot(color="dimgray", notch=T) +
  geom_point(position="jitter", # identity -> default
             color="skyblue", size=2, alpha=0.5)
````

````r
ggplot(Salaries, aes(x=rank, y=salary)) +
  geom_boxplot(color="dimgray", notch=T) +
  geom_point(position="jitter", # identity -> default
             color="skyblue", size=2, alpha=0.5) +
  geom_rug(sides="lr", alpha=0.6) # t b l r
````
