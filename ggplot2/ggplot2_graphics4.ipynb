````r
ggplot(Salaries, aes(x=rank, y=salary,
                     fill=sex)) +
    geom_boxplot()
````

````r
unique(Salaries['rank'])
unique(Salaries$rank)
````

````r
table(Salaries['rank'])
table(Salaries$rank)
````

````r
ggplot(Salaries, aes(x=rank, y=salary,
                     fill=sex)) +
    geom_boxplot() +
    scale_x_discrete(breaks=c("AsstProf", "AssocProf", "Prof"),
                     labels=c("Assistanct\nProfessor", "Associate\nProfessor", "Professor"))
````

````r
ggplot(Salaries, aes(x=rank, y=salary,
                     fill=sex)) +
    geom_boxplot() +
    scale_x_discrete(breaks=c("AsstProf", "AssocProf", "Prof"),
  # labels 에 지정 안할 시 breaks 꺼로 사용
                     labels=c("Assistanct\nProfessor", "Associate\nProfessor", "Professor")) +
    scale_y_continuous(breaks=c(50000, 100000, 150000, 200000),
                       labels=c("$50k", "$100k", "$150k", "$200k"))
````

````r
scale_x_discrete()
scale_x_continuous()
````

````r
ggplot(Salaries, aes(x=rank, y=salary,
                     fill=sex)) + ####
    geom_boxplot() +
    scale_x_discrete(breaks=c("AsstProf", "AssocProf", "Prof"),

                     labels=c("Assistanct\nProfessor", "Associate\nProfessor", "Professor")) +
    scale_y_continuous(breaks=c(50000, 100000, 150000, 200000),
                       labels=c("$50k", "$100k", "$150k", "$200k")) +
    labs(fill="Gender")
````

````r
ggplot(Salaries, aes(x=rank, y=salary,
                     fill=sex)) + ####
    geom_boxplot() +
    scale_x_discrete(breaks=c("AsstProf", "AssocProf", "Prof"),

                     labels=c("Assistanct\nProfessor", "Associate\nProfessor", "Professor")) +
    scale_y_continuous(breaks=c(50000, 100000, 150000, 200000),
                       labels=c("$50k", "$100k", "$150k", "$200k")) +
    scale_fill_discrete(name="Gender") +
    theme(legend.position="top") # bottom none left right
````

````r
ggplot(Salaries, aes(x=rank, y=salary,
                     fill=sex)) + ####
    geom_boxplot() +
    scale_x_discrete(breaks=c("AsstProf", "AssocProf", "Prof"),

                     labels=c("Assistanct\nProfessor", "Associate\nProfessor", "Professor")) +
    scale_y_continuous(breaks=c(50000, 100000, 150000, 200000),
                       labels=c("$50k", "$100k", "$150k", "$200k")) +
    scale_fill_discrete(name="Gender") +
    theme(legend.position=c(.1, .9))
````

````r
head(mtcars)
tail(mtcars)
````

````r
# factor 로 범주형 변수로 바꾼다.
ggplot(mtcars, aes(x=wt, y=mpg,
                   shape=factor(cyl), color=factor(cyl))) +
  geom_point()
````

````r
# factor 로 범주형 변수로 바꾼다.
ggplot(mtcars, aes(x=wt, y=mpg,
                   shape=factor(cyl), color=factor(cyl))) + ###
  geom_point() +
  labs(shape="Cylinder")
````

````r
########
# labels
########
ggplot(mtcars, aes(x=wt, y=mpg,
                   shape=factor(cyl), color=factor(cyl))) + ###
  geom_point() +
  labs(shape="Cylinder", color="Cylinder")
````

````r
#######
# scale
#######
ggplot(mtcars, aes(x=wt, y=mpg,
                   shape=factor(cyl), color=factor(cyl))) + ###
  geom_point() +
  scale_shape_discrete(name="Cylinder") +
  scale_color_discrete(name="Cylinder")
````

````r
ggplot(Salaries, aes(x=rank, fill=sex)) +
  geom_bar() +
  scale_fill_manual(values=c("tomato", "cornflowerblue")) +
  labs(fill="Gender")
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary,
                     color=rank)) +
    geom_point(size=2) +
    scale_color_manual(values=c("orange","violetred","steelblue"))
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary,
                     color=rank)) +
    geom_point(size=2) +
    scale_color_brewer(palette="Accent") # 범주형 일때만.!

    #    Accent    8
    #    Dark2     8
    #    Paired   12
    #    Pastel1   9
    #    Pastel2   8
    #    Set1      9
    #    Set2      8
    #    Set3     12
````
