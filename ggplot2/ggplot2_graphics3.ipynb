````r
install.packages("car") # 5 mins

library(ggplot2)
library(car)
````

````r
head(Salaries)
tail(Salaries)
````

````r
ggplot(Salaries, aes(x=salary)) +
  geom_density()
````

````r
ggplot(Salaries, aes(x=salary, fill=rank)) +
  geom_density(alpha=.5)
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary,
                     color=rank, shape=sex)) +
  geom_point()
````

````r
ggplot(Salaries, aes(x=rank)) +
  geom_bar()
````

````r
ggplot(Salaries, aes(x=rank, fill=sex)) +
  geom_bar(position="stack") # default
````

````r
ggplot(Salaries, aes(x=rank, fill=sex)) +
  geom_bar(position='dodge')
````

````r
ggplot(Salaries, aes(x=rank, fill=sex)) +
  geom_bar(position='fill') +
  labs(y="Propotion")
````

````r
presummed = data.frame(Grade=c(LETTERS[1:5]),
                        Frequency=c(20, 45, 20, 10, 5))
presummed
````

````r
ggplot(presummed, aes(x=Grade, y=Frequency)) +
  geom_bar(stat="identity") # count -> default
````

````r
ggplot(presummed, aes(x=Grade, y=Frequency)) +
  geom_col()
````

````r
library(lattice) # 격자 US∙GB [ ˈlætɪs ]
head(singer)
tail(singer)
````

````r
ggplot(singer, aes(x=height)) +
  geom_histogram() +
  facet_wrap( ~ voice.part, nrow=4) # facet 측면 US∙GB [ ˈfæsɪt' ]
````

````r
ggplot(singer, aes(x=height)) +
  geom_histogram() +
  facet_wrap( ~ voice.part)
````

````r
ggplot(singer, aes(x=height)) +
  geom_density() +
  facet_grid()
````

````r
ggplot(singer, aes(x=height)) +
  geom_density() +
  facet_grid(voice.part ~ .)
````

````r
ggplot(singer, aes(x=height, fill=voice.part)) +
  geom_density() +
  facet_grid(voice.part ~ .) # 한쪽르 생략한다 '.'
````

````r
ggplot(singer, aes(x=height, fill=voice.part)) +
  geom_density() +
  facet_grid(. ~ voice.part) # 한쪽르 생략한다 '.'
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary)) +
  geom_point() +
  facet_grid(sex ~ rank) # 행 ~ 열
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary)) +
  geom_point() +
  facet_grid(. ~ sex) # 행 ~ 열
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary,
                    color=rank, shape=rank)) +
  geom_point() +
  facet_grid( ~ sex) # 행 ~ 열
````
