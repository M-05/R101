````r
library(ggplot2)
library(car)

str(Salaries)
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary, # 박사학위 취득 후 경과 # 연봉
                    color=rank, shape=rank)) +
    geom_point()
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary, # 박사학위 취득 후 경과 # 연봉
                    color=rank, shape=rank)) +
    geom_point() +
    facet_grid(. ~ sex)
  # + theme_gray()
````

````r
ggplot(Salaries, aes(x=yrs.since.phd, y=salary, # 박사학위 취득 후 경과 # 연봉
                    color=rank, shape=rank)) +
    geom_point() +
    facet_grid(. ~ sex) +
    theme_bw()
    # + theme_minimal()
````

````r
ggplot(Salaries, aes(x=rank, y=salary, fill=sex)) +
  geom_boxplot() +
  labs(title="Salary by Rank and Sex",
        x="Rank", y="Salary")
````

````r
ggplot(Salaries, aes(x=rank, y=salary, fill=sex)) +
  geom_boxplot() +
  labs(title="Salary by Rank and Sex",
        x="Rank", y="Salary") +
  theme(plot.title=element_text(face="bold.italic"))

# plot.title: plot title (text appearance) (‘element_text()’; inherits
#           from ‘title’) left-aligned by default
````

````r
ggplot(Salaries, aes(x=rank, y=salary, fill=sex)) +
  geom_boxplot() +
  labs(title="Salary by Rank and Sex",
        x="Rank", y="Salary") +
  theme(plot.title=element_text(face="bold.italic",
                                size=14, color="brown"),
        axis.title=element_text(face="bold.italic",
                                size=10, color="tomato"))
````

````r
ggplot(Salaries, aes(x=rank, y=salary, fill=sex)) +
  geom_boxplot() +
  labs(title="Salary by Rank and Sex",
        x="Rank", y="Salary") +
  theme(axis.text=element_text(face="bold",
                                size=15, color="royalblue"))
````

````r
ggplot(Salaries, aes(x=rank, y=salary, fill=sex)) +
  geom_boxplot() +
  labs(title="Salary by Rank and Sex",
        x="Rank", y="Salary") +
  theme(axis.text=element_text(face="bold",
                                size=15, color="royalblue"),
        panel.background=element_rect(fill="snow", color="darkblue") ,
        panel.grid.major.y=element_line(color='Gray', linetype="solid"))
````

````r
 ggplot(Salaries, aes(x=rank, y=salary, fill=sex)) +
  geom_boxplot() +
  labs(title="Salary by Rank and Sex",
        x="Rank", y="Salary") +
  theme(axis.text=element_text(face="bold",
                                size=15, color="royalblue"),
        panel.background=element_rect(fill="snow", color="darkblue") ,
        panel.grid.major.y=element_line(color='Gray', linetype="solid"),  # 실선
        panel.grid.minor.y=element_line(color='gray', linetype="dashed"), # 점선
        legend.position="top")
````
